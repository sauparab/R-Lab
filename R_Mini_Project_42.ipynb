```r
data <- read.csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv",header=T,sep=";")
data
```

```r
fivenum(data$alcohol)
fivenum(data$chlorides)
```

```r
colMeans(data)
mean(data$fixed.acidity)
plot(data$volatile.acidity,data$sulphates,xlim=c(0,1),ylim=c(0,1))
```

```r
cor(data$volatile.acidity,data$sulphates,method = c("pearson", "kendall", "spearman"))
cor(data$volatile.acidity,data$sulphates,method="pearson")
cor(data$volatile.acidity,data$sulphates,method="kendall")
cor(data$volatile.acidity,data$sulphates,method="spearman")
```

```r
datta <- data.frame(values=c(data$volatile.acidity,data$sulphates),group=c(rep("Volatile acidity",length(data$volatile.acidity)),rep("Sulphates",length(data$sulphates))))
boxplot(values ~ group,datta,col=c("red","yellow"))
datta
install.packages("ggplot2")
library("ggplot2")
ggplot(datta,aes(x=group,y=values,fill=group))+
geom_boxplot()
```
